```python
import os
from dotenv import load_dotenv
from bs4 import BeautifulSoup
import requests
import pandas as pd
import requests
from datetime import datetime
```

```python
load_dotenv()
```

```python
df = pd.read_csv('../CSVs/matches.csv')

def get_season_id(match_date):
    ## 1 = 18/19 season, 2 = 19/20, 3 = 20/21
    if match_date < datetime.date(datetime.strptime('2019-05-13','%Y-%m-%d')):
        return 1
    elif match_date < datetime.date(datetime.strptime('2019-07-27','%Y-%m-%d')):
        return 2
    else:
        return 3
    
df['match_date'] = pd.to_datetime(df['kickoff']).dt.date
df['kickoff'] = pd.to_datetime(df['kickoff']).dt.time
df['season_id'] = df['match_date'].apply(lambda x: get_season_id(x))
df['home_team'] = df['home_team'].str.lower().str.strip()
df['away_team'] = df['away_team'].str.lower().str.strip()
df['referee'] = df['referee'].str.lower().str.strip()
df['stadium_name'] = df['stadium_name'].str.lower().str.strip()
df['stadium_city'] = df['stadium_city'].str.lower().str.strip()
df.to_csv('../CSVs/matches_cleaned.csv',index=False)
df
```
