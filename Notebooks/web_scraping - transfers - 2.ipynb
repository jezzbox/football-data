```python
import os
from dotenv import load_dotenv
from bs4 import BeautifulSoup
import requests
import pandas as pd
import requests
```

```python
load_dotenv()

def get_soup(url):
    page = requests.get(url,headers = {"User-Agent":"Mozilla/5.0","origin":"https://www.transfermarkt.co.uk"})
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup
```

```python
df = pd.read_csv('../CSVs/player_transfers.csv')
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df
```

```python
df["mv"].head(10)
```

```python
import re

def convert_mv(mv):
    if 'm' in str(mv):
        converted_mv = re.sub('[^0-9.]', '', str(mv))
        return float(converted_mv) * 1000000
    elif 'Th.' in str(mv):
        converted_mv = re.sub('[^0-9.]', '', str(mv))
        return float(converted_mv) * 1000
    elif mv == '-' or mv =='?':
        return None
    else:
        return mv
```

```python
df["mv"] = df["mv"].apply(lambda x: convert_mv(x))
df
```

```python
df["fee_details"] = df["fee"]
df["fee"] = df["fee"].apply(lambda x: 0 if x == 'End of loan' or x == 'free transfer' or x == 'loan transfer' else convert_mv(x))
df
```

```python
df.loc[(df["fee_details"] != 'End of loan') & (df["fee_details"] != 'free transfer') & (df["fee_details"] != 'loan transfer'),'fee_details'] = None
df
```

```python
df[["first_name","surname","other_names"]] = df["name"].str.lower().str.split(' ',2,expand=True)
df
```

```python
df = df[["season","date","left","joined","mv","fee","fee_details","first_name","surname","other_names"]]
df.to_csv('../CSVs/player_transfers_cleaned.csv')
```
