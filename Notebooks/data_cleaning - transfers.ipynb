```python
import os
from dotenv import load_dotenv
from bs4 import BeautifulSoup
import requests
import pandas as pd
import requests
```

```python
load_dotenv()
```

```python
df = pd.read_csv('../CSVs/transfers_complete.csv')
df.rename(columns={'Unnamed: 0':'player_transfer_index'},inplace=True)
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df
```

```python
df["mv"].head(10)
```

```python
import re

def convert_mv(mv):
    if 'm' in str(mv):
        converted_mv = re.sub('[^0-9.]', '', str(mv))
        return float(converted_mv) * 1000000
    elif 'Th.' in str(mv):
        converted_mv = re.sub('[^0-9.]', '', str(mv))
        return float(converted_mv) * 1000
    elif mv == '-' or mv =='?':
        return None
    else:
        return mv
```

```python
df["mv"] = df["mv"].apply(lambda x: convert_mv(x))
df
```

```python
df["fee_details"] = df["fee"]
df["fee"] = df["fee"].apply(lambda x: 0 if x == 'End of loan' or x == 'free transfer' or x == 'loan transfer' else convert_mv(x))
df
```

```python
df.loc[(df["fee_details"] != 'End of loan') & (df["fee_details"] != 'free transfer') & (df["fee_details"] != 'loan transfer'),'fee_details'] = None
df
```

```python
df[["first_name","surname","other_names"]] = df["name"].str.lower().str.split(' ',2,expand=True)
```

```python
df["date_from"] = df['date']
df['date_to'] = df['date'].shift(1)
df["contract_expires"] = pd.to_datetime(df["contract_expires"],errors='coerce')
df
```

```python
df.loc[df["player_transfer_index"] == 0,"date_to"] = df["contract_expires"]
df
```

```python
df["dob"] = pd.to_datetime(df["dob"])
df
```

```python
df["left"] = df["left"].str.lower().str.strip()
df["joined"] = df["joined"].str.lower().str.strip()
df["fee_details"] = df["fee_details"].str.lower().str.strip()
df["city"] = df["city"].str.lower().str.strip()
df["country"] = df["country"].str.lower().str.strip()
df["height"] = df["height"].str.replace(",",".").str.replace("m","").str.strip()
df = df[["season","date_from","date_to","left","joined","mv","fee","fee_details","first_name","surname","other_names","dob","city","country","height","position"]].drop_duplicates()
df
```

```python
df.to_csv('../CSVs/transfers_cleaned.csv')
```
